## Veri HazÄ±rlÄ±ÄŸÄ±

### 1. StringIndexer

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoderEstimator, VectorAssembler, StandardScaler
```

```python
country_indexer = StringIndexer() \
.setInputCol('Country') \
.setOutputCol('country_index') \
.setHandleInvalid('skip')
```

```python
status_indexer = StringIndexer() \
.setInputCol('Status') \
.setOutputCol('status_index') \
.setHandleInvalid('skip')
```

### 2. OneHotEncoder

```python
encoder = OneHotEncoderEstimator() \
.setInputCols(['status_index']) \
.setOutputCols(['status_encoded'])
```

### 3. VectorAssembler

```python
numerik_nitelikler = ['Year', 'Adult Mortality', 'infant deaths', 'Alcohol', 'BMI', 'under-five deaths', 'Total expenditure', 
                      'HIV/AIDS', 'Income composition of resources', 'Schooling', 'status_encoded']
```

```python
vector_assembler = VectorAssembler() \
.setInputCols(numerik_nitelikler) \
.setOutputCol('vectorized_features')
```

### 4. StandardScaler

```python
scaler = StandardScaler() \
.setInputCol('vectorized_features') \
.setOutputCol('features')
```

### 5. Linear Regression

```python
from pyspark.ml.regression import LinearRegression
```

```python
lr_object = LinearRegression() \
.setFeaturesCol('features') \
.setLabelCol('label') \
.setPredictionCol('prediction')
```

### 6. Train-Test Split

```python
train_df, test_df = df_prep2.randomSplit([0.8, 0.2], seed = 142)
```

### 7. Pipeline

```python
from pyspark.ml import Pipeline
```

```python
pipeline_nesnesi = Pipeline() \
.setStages([status_indexer, encoder, vector_assembler, scaler, lr_object])
```

### 8. Modeli EÄŸitme

```python
pipeline_model = pipeline_nesnesi.fit(train_df)
```

```python
pipeline_model.transform(test_df).select('label', 'prediction').toPandas().head(10)
```
